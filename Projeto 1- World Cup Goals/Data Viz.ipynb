### Separação dos paises e gols.
Nós precisamos dos paises, quantidade de gols e ano.
Como está dividido em home_team e away_team, precisamos acabar com essa separação, pois será necessários para somarmos a quantidade de gols total que um país fez.

País da casa

```python
home_score = results_fifa[['home_team','home_score','year']]
home_score
```

País visitante

```python
away_score = results_fifa[['away_team','away_score','year']]
away_score
```

- Para juntarmos as duas tabelas precisamos modificar o nome das colunas e deixá-los iguais.\
Para isso usamos o 'rename'.

País da casa

```python
home_score = results_fifa[['home_team','home_score','year']].rename({'home_team':'team','home_score':'score'},axis=1)
home_score
```

País visitante

```python
away_score = results_fifa[['away_team','away_score','year']].rename({'away_team':'team','away_score':'score'},axis=1)
away_score
```

- Agora que as planilhas estão com nomes iguais, podemos uni-las com o 'concat'.

```python
score_fifa = pd.concat([home_score,away_score])
score_fifa
```

- Ajustando o index

```python
score_fifa.reset_index(drop=True,inplace=True)
```

```python
score_fifa
```

- Agrupando o numero de gols por time e por ano.

```python
score_fifa.groupby(['team','year']).score.sum()
```

- Fazendo a mesma coisa que fiz acima, porém transformando em um DataFrame para visualizar melhor.

```python
pd.DataFrame(score_fifa.groupby(['team','year']).score.sum())
```

- Pivotando os dados da planilha para ficar da meneira correta de visualizar.

```python
score_fifa.groupby(['team','year']).score.sum().unstack()
```

- Como pivotamos a planilha, apareceu uma quantidade enorme de valores NaN (Not a Number), pois os países que não participaram da copa no ano ou não fizeram gol vai aparecere que não existe dado, já na forma que estávamos visualizando antes os anos que as seleções não marcavam gols ou não participaram da copa não aparecem.

- Usamos um 'fillna' para resolver isso.

```python
score_fifa.groupby(['team','year']).score.sum().unstack().fillna(0)
```

```python
score_fifa.groupby(['team','year']).score.sum().unstack().fillna(0).cumsum(axis=1).reset_index()
```

```python
score_fifa_year = score_fifa.groupby(['team','year']).score.sum().unstack().fillna(0).cumsum(axis=1).reset_index()
score_fifa_year
```

- Importando os dados das bandeiras, confederações e siglas dos países.

```python
flags = pd.read_csv('countries-fifa-flags.csv')
flags
```

- Unindo dois DataFrames lateralmente como 'merge'.

```python
pd.merge(right=flags,left=score_fifa_year,left_on='team',right_on='country')
```

- Analisamos qu nem todas as linhas foram para o novo DF.

Vamos analisar quais linhas não foram adicionadas com a função 'isin'.

```python
score_fifa_year[~score_fifa_year['team'].isin(flags.country)]
```

- Para podermos colocar as colunas ignorando o erro gerado pela diferença entre os nomes de flags.country e score_fifa_year.team, vamos usar o parâmetro 'how' e indicar qual DataFrame queremos usar como fixo.

```python
dados = pd.merge(right=flags,left=score_fifa_year,left_on='team',right_on='country', how='left')
dados
```

- Vamos salvar o nosso DF filtrado, organizado e ajustado para mandar par ao Flourish

```python
dados.to_csv('fifa_data_viz.csv',index=False)
```
